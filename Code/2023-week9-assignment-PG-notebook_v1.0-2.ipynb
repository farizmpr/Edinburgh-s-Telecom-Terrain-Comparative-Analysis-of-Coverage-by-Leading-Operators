```python
def fetch_data(min_lat, max_lat, min_long, max_long, radio):
    global fetched_data_df

    # Check if data for the current selection is already fetched
    if not fetched_data_df.empty and (fetched_data_df['radio'].iloc[0] == radio):
        return fetched_data_df
    try:
        api_key = "pk.899971c4aaf9df677b0e5a73decb418c"
        payload = f"https://www.opencellid.org/cell/getInArea?key={api_key}&BBOX={min_lat},{min_long},{max_lat},{max_long}&format=json&radio={radio}"
        data = requests.get(payload).json()
        if(len(data) > 0):
            fetched_data_df = pd.DataFrame(data['cells'])
            fetched_data_df['radio'] = radio  
        else:
            print("Error! No data found")
    except:
        print("Error fetching data")
        print(data)

    return fetched_data_df
```

```python
def analyze_network(area, radio, network):
    area_coords = define_areas()[area]
    latitudes, longitudes = zip(*area_coords)
    min_lat, max_lat, min_long, max_long = min(latitudes), max(latitudes), min(longitudes), max(longitudes)

    df_data = fetch_data(min_lat, max_lat, min_long, max_long, radio)

    # Filter data by polygon
    polygon = Polygon(area_coords)
    df_data['point'] = df_data.apply(lambda row: Point(row['lat'], row['lon']), axis=1)
    df_filtered = df_data[df_data['point'].apply(polygon.contains)]

    # Enrich data by adding Network name
    df_mnc_lookup = pd.DataFrame({'mnc': [30, 20, 10, 15], 'Network': ['EE', '3', 'O2', 'Vodafone']})
    df_enriched = pd.merge(df_filtered, df_mnc_lookup, on='mnc', how='inner')

    if network != 'All':
        df_enriched = df_enriched[df_enriched['Network'] == network]
        
    # Analyze coverage
    df_enriched['lat_bin'] = df_enriched['lat'].apply(lambda latitude: float(str(latitude)[:6]))
    df_enriched['lon_bin'] = df_enriched['lon'].apply(lambda longitude: float(str(longitude)[:6]))
    df_summary = df_enriched.groupby(['Network', 'lat_bin', 'lon_bin']).size().reset_index(name='counts')

    return df_summary
```

```python
def visualize_coverage(area, df_summary=None):

    if df_summary is not None and not df_summary.empty:
        # Center map on the selected area
        center_lat = sum(lat for lat, _ in define_areas()[area]) / len(define_areas()[area])
        center_lon = sum(lon for _, lon in define_areas()[area]) / len(define_areas()[area])
        zoom_start = 15
    else:
        center_lat, center_lon = 55.9500, -3.2000  
        zoom_start = 14 

    heatmap_map = folium.Map(location=[center_lat, center_lon], zoom_start=zoom_start)

    # Add heat layers if data is provided
    if df_summary is not None and not df_summary.empty:
        for network in df_summary['Network'].unique():
            network_data = df_summary[df_summary['Network'] == network]
            heat_data = network_data[['lat_bin', 'lon_bin', 'counts']].values.tolist()
            heat_layer = HeatMap(heat_data, name=network)
            heat_layer.add_to(heatmap_map)
        folium.LayerControl().add_to(heatmap_map)

    display(heatmap_map)
```

```python
def on_selection_change(change):
    global selected_area, selected_radio, selected_network, fetched_data_df
    clear_output(wait=True)

    # Check if area or radio type has changed
    area_or_radio_changed = False
    if change['new'] in define_areas() and selected_area != change['new']:
        selected_area = change['new']
        area_or_radio_changed = True
    elif change['new'] in ['UMTS', 'LTE', 'GSM'] and selected_radio != change['new']:
        selected_radio = change['new']
        area_or_radio_changed = True
    elif change['new'] in ['All', 'EE', '3', 'O2', 'Vodafone']:
        selected_network = change['new']

    # Reset fetched data if area or radio type changed
    if area_or_radio_changed:
        fetched_data_df = pd.DataFrame()

    initialize_user_interface(selected_area, selected_radio, selected_network)
    df_summary = analyze_network(selected_area, selected_radio, selected_network)
    visualize_coverage(selected_area, df_summary)
```

```python
def initialize_user_interface(current_area, current_radio, current_network):
    area_selection_widget = widgets.ToggleButtons(
        options=list(define_areas().keys()),
        value=current_area,
        description='Select area',
        disabled=False,
        button_style=''
    )

    radio_selection_widget = widgets.ToggleButtons(
        options=['GSM', 'UMTS', 'LTE'],
        value=current_radio,
        description='Select radio',
        disabled=False,
        button_style=''
    )

    network_selection_widget = widgets.ToggleButtons(
        options=['All', 'EE', '3', 'O2', 'Vodafone'],
        value=current_network,
        description='Select network',
        disabled=False,
        button_style=''
    )

    area_selection_widget.observe(on_selection_change, names='value')
    radio_selection_widget.observe(on_selection_change, names='value')
    network_selection_widget.observe(on_selection_change, names='value')

    display(area_selection_widget, radio_selection_widget, network_selection_widget)
```
